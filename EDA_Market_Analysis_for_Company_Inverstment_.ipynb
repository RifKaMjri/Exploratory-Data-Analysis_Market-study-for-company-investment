# Hypothesis1
H0 : There is no difference regarding Age for Yellow and Pink Cab

H1 : There is difference regarding Age for both Yellow and Pink Cab

```python
#Pink cab
First_dataset = Merged_data[(Merged_data.age<=50)&(Merged_data.company=='Pink Cab')].groupby('transaction_id').Profit.mean()
Second_dataset = Merged_data[(Merged_data.age>50)&(Merged_data.company=='Pink Cab')].groupby('transaction_id').Profit.mean()
print(First_dataset.shape[0],Second_dataset.shape[0])

from scipy import stats
_, p_value = stats.ttest_ind(First_dataset.values, Second_dataset.values,equal_var=True)
if(p_value<0.05):
    print('We accept alternate hypothesis that theres a difference')
else:
    print('We accept null hypothesis that theres no difference')
    
print('P value is ', p_value)
```

```python
#Yellow Cab

First_dataset = Merged_data[(Merged_data.age<=50)&(Merged_data.company=='Yellow Cab')].groupby('transaction_id').Profit.mean()
Second_dataset = Merged_data[(Merged_data.age>50)&(Merged_data.company=='Yellow Cab')].groupby('transaction_id').Profit.mean()
print(First_dataset.shape[0],Second_dataset.shape[0])

from scipy import stats
_, p_value = stats.ttest_ind(First_dataset.values, Second_dataset.values,equal_var=True)
if(p_value<0.05):
    print('We accept alternate hypothesis that theres a difference')
else:
    print('We accept null hypothesis that theres no difference')
    
print('P value is ', p_value)
```

```python
#Conduct t-test to measure the difference between Trasaction and Gender in both companies
#Yellow Cab 
Data_female = Merged_data[(Merged_data.gender=='Female')&(Merged_data.company=='Yellow Cab')].groupby('transaction_id').Profit.mean()
Data_Male = Merged_data[(Merged_data.gender=='Male')&(Merged_data.company=='Yellow Cab')].groupby('transaction_id').Profit.mean()
print(Data_female.shape[0],Data_Male.shape[0])


#import scipy to conduct testing hypothesis T-test
from scipy import stats

t, p_value = stats.ttest_ind(a = Data_female.values,b=Data_Male.values,equal_var=True)
if(p_value<0.05):
    print('We accept alternate hypothesis that there is a statistical difference in Yellow Cab between gender ')
else:
    print('We accept null hypothesis that there is no statistical difference in Yellow Cab between gender')
    
print('P value is ', p_value)


#Pink Cab
Data_female = Merged_data[(Merged_data.gender=='Female')&(Merged_data.company=='Pink Cab')].groupby('transaction_id').Profit.mean()
Data_Male = Merged_data[(Merged_data.gender=='Male')&(Merged_data.company=='Pink Cab')].groupby('transaction_id').Profit.mean()
print(Data_female.shape[0],Data_Male.shape[0])

t, p_value = stats.ttest_ind(a = Data_female.values,b=Data_Male.values,equal_var=True)
if(p_value<0.05):
    print('We accept alternate hypothesis that there is a statistical difference in Pink Cab between gender')
else:
    print('We accept null hypothesis that there is no statistical difference in Pink Cab between gender')
    
print('P value is ', p_value)
```
